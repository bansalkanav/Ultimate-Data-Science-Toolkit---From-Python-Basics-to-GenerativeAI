# Data Distributions
A data distribution refers to the way in which data values are spread or distributed across a dataset. It provides information about the possible values a variable can take and how frequently each value occurs. Understanding the distribution of data is fundamental in statistics and data analysis, as it forms the basis for making inferences, building models, and drawing conclusions about a population based on a sample.

Key characteristics of a data distribution include:
1. **Central Tendency:** This indicates the "center" of the data. Common measures of central tendency include the mean (average), median (middle value), and mode (most frequently occurring value).

2. **Dispersion or Spread:** This describes how spread out the data points are. Measures of dispersion include range, variance, and standard deviation.

3. **Shape:** This refers to the overall pattern or form of the distribution. Common shapes include normal (bell-shaped), skewed (lopsided), and uniform (evenly distributed).

4. **Outliers:** Outliers are data points that are significantly different from the rest of the data. Understanding the distribution helps in identifying and handling outliers appropriately.

5. **Probability Distributions:** In probability theory and statistics, distributions describe the likelihood of different outcomes in a random process. Examples include the normal distribution, binomial distribution, and Poisson distribution.

6. **Cumulative Distribution Function (CDF):** The CDF gives the probability that a random variable takes on a value less than or equal to a specific value.

Understanding the distribution of data is crucial in a wide range of fields, including statistics, data science, economics, finance, healthcare, and more. It informs decision-making, helps in selecting appropriate statistical tests or models, and provides insights into the underlying patterns and characteristics of a dataset.


### Example 1: Exam Scores

Suppose you are a data scientist working in the education domain. You have a dataset containing the final exam scores of 1000 students in a class.

Now, let's explore how understanding distributions can be important in this scenario:

1. **Descriptive Statistics:**  
Mean: Knowing the distribution helps you understand the average score. If the data follows a normal distribution, the mean can provide a good representation of the central tendency.
Standard Deviation: It gives a measure of the spread of the data. If the data is normally distributed, about 68% of the data falls within one standard deviation of the mean.

2. **Inferential Statistics:**  
Hypothesis Testing: If you want to compare the scores of two groups (e.g., male vs. female students), understanding the underlying distribution is crucial for choosing the right statistical test (e.g., t-test for normal distributions).
Confidence Intervals: Understanding the distribution allows you to compute confidence intervals, which provide a range of values within which you can be reasonably confident the true parameter lies.

3. **Modeling:**  
Regression Analysis: If you want to predict future exam scores based on other factors (e.g., study hours, attendance), knowing the distribution of the target variable helps in selecting appropriate regression techniques.
Classification: For binary classification problems, understanding the distribution of classes helps in evaluating the model's performance (accuracy, precision, recall, etc.).

4. **Outlier Detection:**  
Knowing the distribution can help you identify outliers. For instance, if the data is normally distributed, you can use z-scores to detect outliers.

5. **Decision Making:**
If the scores follow a specific distribution, you can set grading curves or define cutoffs for different grades based on percentiles.

6. **Data Generation:**
If you need to generate synthetic data for testing or simulations, understanding the distribution of the original data can guide you in generating new samples.

In summary, understanding the distribution of data is fundamental in various aspects of data analysis, from descriptive statistics to modeling and decision-making. It enables data scientists to draw meaningful insights and make informed decisions based on the characteristics of the data at hand.

## Plotting Uniform and Normal Distribution

```python
uniform_100 = np.random.uniform(low = 10, high = 40, size = (100))

sns.distplot(uniform_100, kde=True)

plt.show()
```

```python
uniform_1000 = np.random.uniform(low = 10, high = 40, size = (1000))

sns.distplot(uniform_1000, kde=True)

plt.show()
```

```python
uniform_10000 = np.random.uniform(low = 10, high = 40, size = (10000))

sns.distplot(uniform_10000, kde=True)

plt.show()
```

```python
normal_100 = np.random.normal(loc = 20, scale = 5, size=100)

sns.distplot(normal_100, kde=True)

plt.show()
```

```python
normal_1000 = np.random.normal(loc = 20, scale = 5, size=1000)

sns.distplot(normal_1000, kde=True)

plt.show()
```

```python
normal_10000 = np.random.normal(loc = 20, scale = 5, size=10000)

sns.distplot(normal_10000, kde=True)

plt.show()
```

```python
mu_normal_10000 = normal_10000.mean()

sigma_normal_10000 = normal_10000.std()

print(mu_normal_10000, sigma_normal_10000)
```

```python
one_std_right = mu_normal_10000 + (1 * sigma_normal_10000)

one_std_left = mu_normal_10000 - (1 * sigma_normal_10000)

two_std_right = mu_normal_10000 + (2 * sigma_normal_10000)

two_std_left = mu_normal_10000 - (2 * sigma_normal_10000)

three_std_right = mu_normal_10000 + (3 * sigma_normal_10000)

three_std_left = mu_normal_10000 - (3 * sigma_normal_10000)
```

```python
plt.figure(figsize=(20,10))

plt.axvline(mu_normal_10000, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')

plt.legend()

sns.set_style("darkgrid")
sns.distplot(normal_10000)

plt.show()
```

## Checking Skewness and Kurtosis

### Skewness

In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:

- If skewness is less than -1 or greater than 1, the distribution is highly skewed.
- If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
- If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.


### Kurtosis

Kurtosis tells you the height and sharpness of the central peak, relative to that of a standard bell curve.
Kurtosis is a measure of the tailedness of a distribution. Tailedness is how often outliers occur. 
Excess kurtosis is the tailedness of a distribution relative to a normal distribution. 

Distributions with medium kurtosis (medium tails) are mesokurtic. Mesokurtic distributions have outliers that are neither highly frequent, nor highly infrequent.
Distributions with low kurtosis (thin tails) are platykurtic. Uniform distributions are platykurtic. Platykurtic distributions have a low frequency of outliers. 
Distributions with high kurtosis (fat tails) are leptokurtic. Leptokurtic distributions have frequent outliers. 

**Question - What is the difference between skewness and kurtosis?**  
Skewness and kurtosis are both important measures of a distributions shape.
- Skewness measures the asymmetry of a distribution.
- Kurtosis measures the heaviness of a distributions tails relative to a normal distribution.

**Question - What is a normal distribution?**  
In a normal distribution, data are symmetrically distributed with no skew. Most values cluster around a central region, with values tapering off as they go further away from the center.

The measures of central tendency (mean, mode, and median) are exactly the same in a normal distribution.

## Box - Cox Transformation

```python
pareto_rv = np.random.pareto(a = 2, size = 1000)

sns.distplot(pareto_rv)
```

```python
# Normality Test

stats.probplot(pareto_rv, dist="norm", plot=plt)

plt.grid()
```

```python
# x_t = tranformed by box-cox, l = Lambda

x_t, l = stats.boxcox(pareto_rv)

print(l)
```
