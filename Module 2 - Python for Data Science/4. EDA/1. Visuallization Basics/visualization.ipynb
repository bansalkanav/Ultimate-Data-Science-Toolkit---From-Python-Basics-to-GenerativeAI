# Basic Plotting with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)

y = x ** 2

plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y, 'b+')

plt.show()
```

```python
# Labeling

plt.plot(x, y, 'bo', x, y, 'k-')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot - x & y')

plt.xlim([0, 10])
plt.ylim([0, 100])

plt.show()
```

```python
plt.plot([1, 3, 5, 7, 9], [1, 8, 3, 5, 1])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot - x & y')

plt.xlim([0, 10])
plt.ylim([0, 10])

plt.show()
```

```python
# Labeling the plot

heights = np.array([160, 180, 146, 162, 184, 180])

weights = np.array([50, 78, 45, 51, 80, 60])

plt.plot(heights, weights, 'ko')

plt.xlabel('Heights (cm)')
plt.ylabel('Weights (kg)')
plt.title('Scatter Plot - Height & Weight')

plt.xlim([140, 200])
plt.ylim([40, 90])

plt.show()
```

```python
# Multiple plots in a single figure

x = np.arange(1, 21)
y_1 = np.log(x)
y_2 = np.exp(x)

# Initiating a figure
plt.figure()

# Creating a plot with 1 row and 2 cols

plt.subplot(1, 2, 1)
plt.title('y = log(x)')
plt.plot(x, y_1)

plt.subplot(1, 2, 2)
plt.title('y = exp(x)')
plt.plot(x, y_2)

plt.show()
```

# Plotting with Seaborn

## 1 Univariate Analysis (for Numerical Variables)
i.e Analyzing one variable at a time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/random.csv')

print(df.shape)

df.head()
```

```python
sns.distplot(df['y'])

# by default-> sns.distplot(x, hist=True, kde=True)
```

### 1.1 Histogram

<b>REFERENCE: </b>https://statistics.laerd.com/statistical-guides/understanding-histograms.php

```python
sns.distplot(df['y'], kde=False, rug=True)

# This is basically a frequency plot
# Try 'rug=False'
```

### 1.2 Kernel Density Estimation (1D Dimensionality plot)

<b>REFERENCE: </b>http://www.mvstat.net/tduong/research/seminars/seminar-2001-05/

```python
sns.distplot(df['y'], hist=False, rug=True)
```

## 3 Working With Categorical Data

```python
tips = pd.read_csv('data/tips.csv')

tips.head()
```

```python
tips.shape
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

### 3.1 Categorical Scatterplots - stripplot()

```python
sns.stripplot(data = tips, x = 'day', y='total_bill')
```

### 3.2 Swarm Plot

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a beeswarm.

```python
sns.swarmplot(data = tips, x='day', y='total_bill')
```

### 3.3 Boxplot

```python
sns.boxplot(data = tips, x='day', y='total_bill')
```

```python
tips.loc[tips.time == 'Lunch', :]
```

```python
tips.loc[tips.time == 'Lunch', 'day'].value_counts()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
```

```python
# Adding a new column, indicating weekend or not

tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])

tips.head()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='weekend')
```
