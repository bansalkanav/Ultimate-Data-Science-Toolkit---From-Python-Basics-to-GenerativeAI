## Anatomy of a Figure
<img style="float: right;" width="400" height="400" src="images/00_anatomy_of_plot.PNG">

Matplotlib graphs your data on <b>Figures</b> (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more <b>Axes</b>, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.).

Below mentioned are the parts of Figure:
- Figure - The whole Figure.
- Axes - An <b>Axes</b> is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects (<b>be aware of the difference between Axes and Axis</b>) that provide ticks and tick labels to provide scales for the data in the Axes. Each Axes also has a title (set via <b>set_title()</b>), an x-label (set via <b>set_xlabel()</b>), and a y-label set via <b>set_ylabel()</b>)
- Artist - Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects). This includes Text objects, Line2D objects, collections objects, Patch objects, etc. When the Figure is rendered, all of the Artists are drawn to the canvas. Most Artists are tied to an Axes; such an Artist cannot be shared by multiple Axes, or moved from one to another.

### Labelling the Plots

````python
# Initialize a figure and fix the size to width=3inches and height=3inches
plt.figure(figsize=(3,3))

# Generate scatter and line plots between data1 and data2 
plt.plot(data1, data2, 'bo', label="scatter")
plt.plot(data1, data2, 'k-', label="line")

# Labelling the plot
plt.title('Line Plot')
plt.xlabel('x label')
plt.ylabel('y label')

# Putting limits to x and y axis
plt.xlim([0, 1])
plt.ylim([0, 1])

# Rotate x ticks
plt.xticks(rotation=45)

# Display the ledgend
plt.legend()

# Display the Grid
plt.grid(True)

# Display the figure
plt.show()
````

### Annotating the Plots (Text and Arrows)

````python
# Initialize a figure and fix the size to width=3inches and height=3inches
plt.figure(figsize=(3,3))

# Generate scatter and line plots between data1 and data2 
plt.plot(data1, data2, 'bo', label="scatter")
plt.plot(data1, data2, 'k-', label="line")

# Labelling the plot
plt.title('Line Plot')
plt.xlabel('x label')
plt.ylabel('y label')

# Annotating the plot
plt.text(s="My Text", x=0.5, y=0.6)

# Putting limits to x and y axis
plt.xlim([0, 1])
plt.ylim([0, 1])

# Display the ledgend
plt.legend()

# Display the figure
plt.show()
````

<b>Legend Location</b>

The bbox_to_anchor keyword gives a great degree of control for manual legend placement. For example, if you want your axes legend located at the figure's top right-hand corner instead of the axes' corner, simply specify the corner's location and the coordinate system of that location:

````python
# Initialize a figure and fix the size to width=3inches and height=3inches
plt.figure(figsize=(5, 2.7), layout='constrained')

# Generate line plots
plt.plot(data1, data1, label='linear', color="green")
plt.plot(data1, data1**2, label='quadratic', color='orange')
plt.plot(data1, data1**3, label='cubic', color="blue")

# Labelling the plot
plt.title("Simple Plot")
plt.xlabel('x label')
plt.ylabel('y label')

# Annotating the plot
plt.text(s="cubic", x=1.75, y=5.2)
plt.text(s="quadratic", x=1.75, y=3.5)
plt.text(s="linear", x=1.75, y=2)

# Display the Legend outside the axes
plt.legend(bbox_to_anchor=(1, 1))

# Display the figure
plt.show()
````

### Labelling and Annotating Subplots

````python
# SYNTAX 1
# Initializing a figure (without axes)
fig = plt.figure(figsize=(3, 3), facecolor='lightskyblue')
fig.suptitle("Figure Title")

# Adding Axes (Generates 1 axes or subplot)
ax = fig.subplots()
ax.set_title("Axes Title", loc="left", fontsize="small")

# Adding Annotations
ax.annotate("My Text", xytext=(0.2, 0.2), ha="center",
            xy=(0.5, 0.5), arrowprops={})

# Display the Figure
plt.show()
````

````python
# SYNTAX 2
# Initializing a figure (with 1 axes)
fig, ax = plt.subplots(figsize=(3, 3), facecolor='lightskyblue')
fig.suptitle("Figure Title")

# Adding Axes Title
ax.set_title("Axes Title", loc="left", fontsize="small")

# Adding Axes Grid
ax.grid(True)

# Adding Axes Color
ax.set_facecolor("yellow")

# Adding Annotations
ax.annotate("My Text", xytext=(0.2, 0.2), ha="center",
            xy=(0.5, 0.5), arrowprops={})

# Display the Figure
plt.show()
````

````python
# Initializing a Figure with 4 Axes (2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                        facecolor='lightskyblue', 
                        layout='constrained')

# Set the color of axes to YELLOW
axs.set_facecolor('yellow')

plt.show()
````

````python
# Initializing a Figure with 4 Axes (2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                        facecolor='lightskyblue', 
                        layout='constrained')

# Since axs is a numpy nd-array, we can iterate and make changes to each subplot
for ax in axs.ravel():
    ax.set_facecolor('yellow')

plt.show()
````

````python
# Initialize the figure with 4 axes (arranged in 2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                   facecolor='lightskyblue', 
                   layout='constrained')
fig.suptitle('Figure')

title  = ['Axes_1', 'Axes_2', 'Axes_3', 'Axes_4']
color = ['yellow', 'violet', 'green', 'pink']

# Iterating over each axes
for idx, ax in enumerate(axs.ravel()):
    ax.set_facecolor(color[idx])
    ax.set_title(title[idx], loc='center', fontsize='medium')
    ax.set_xlabel("x Label")
    ax.set_ylabel("y Label")

plt.show()
````

## The implicit and the explicit interfaces

As noted above, there are essentially two ways to use Matplotlib:
- OO-style - Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style").
- pyplot-style - Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.

Matplotlib's documentation and examples use both the OO and the pyplot styles. In general, matplotlib suggest using the OO style, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. However, the pyplot style can be very convenient for quick interactive work.

````python
plt.figure(figsize=(5, 2.7), layout='constrained')

plt.plot(data1, data1, label='linear')
plt.plot(data1, data1**2, label='quadratic')
plt.plot(data1, data1**3, label='cubic')

plt.title("Simple Plot")
plt.xlabel('x label')
plt.ylabel('y label')

plt.text(x=1.75, y=5.2, s="cubic", color='green')
plt.text(x=1.75, y=3.5, s="quadratic", color='orange')
plt.text(x=1.75, y=2, s="linear", color='blue')

plt.legend()

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax.plot(data1, data1, label='linear', color='green')  # Plot some data on the axes
ax.plot(data1, data1**2, label='quadratic', color='orange')  # Plot more data on the axes...
ax.plot(data1, data1**3, label='cubic', color='blue')  # ... and some more

ax.set_title("Simple Plot")  # Add a title to the axes
ax.set_xlabel('x label')  # Add an x-label to the axes
ax.set_ylabel('y label')  # Add a y-label to the axes

ax.text(s="linear", x=1.75, y=2, color='green')
ax.text(s="quadratic", x=1.75, y=3.5, color='orange')
ax.text(s="cubic", x=1.75, y=5.2, color='blue')


ax.legend()  # Add a legend

plt.show()
````

````python
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout='constrained')
fig.suptitle("Comparision")

data = [data1, data1**2, data1**3]
labels = ['linear', 'quadratic', 'cubic']
colors = ['green', 'orange', 'blue']

for idx, ax in enumerate(axs.ravel()):
    ax.plot(data[0], data[idx], 
            label=labels[idx], 
            color=colors[idx])  # Plot some data on the axes
    
    # Add a horizontal and vertical line
    ax.axhline(1, ls='--', color='red')
    ax.axvline(1, ls='--', color='black')
    
    ax.set_title("Simple Plot")  # Add a title to the axes
    ax.set_xlabel('x label')  # Add an x-label to the axes
    ax.set_ylabel('y label')  # Add a y-label to the axes
        
    ax.legend()  # Add a legend

plt.show()
````

## Using Helper Methods

````python
def my_plotter(axes, data1, data2, kind):
    """
    A helper function to make a graph.
    """
    if kind=="line":
        out = axes.plot(data1, data2)
    else:
        out = axes.scatter(data1, data2)
    return out

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))

my_plotter(ax1, data1, data2, kind="line")
my_plotter(ax2, data3, data4, kind="scatter")

plt.show()
````

## Basic Matplotlib Plots

````python
# Make 4 random data sets

np.random.seed(2)
data1 = np.linspace(0, 100, 10000)
data2 = np.random.randint(1, 6, size=(10000, ))
data3, data4 = np.random.randn(2, 10000)

df = pd.DataFrame({"col_1": data1, 
                   "col_2": data2, 
                   "col_3": data3, 
                   "col_4": data4})

df.head()
````

### Univariate Plotting For Numerical Columns - Histogram and Box Plot

````python
fig, axs = plt.subplots(1, 2, figsize=(6, 3), layout="constrained")
fig.suptitle("Univariate Plotting - Numerical Features")

axs[0].hist(df["col_3"])
axs[0].set_title("Histogram Plot", fontsize="medium")

axs[1].boxplot(df["col_3"])
axs[1].set_title("Box Plot", fontsize="medium")

plt.show()
````

### Univariate Plotting For Categorical Columns - Bar Plot and Pie Chart

````python
fig, axs = plt.subplots(1, 2, figsize=(6, 3), layout="constrained")
fig.suptitle("Univariate Plotting - Categorical Features")

bar_plt_data = df["col_2"].value_counts()
axs[0].bar(bar_plt_data.index, bar_plt_data.values)
axs[0].set_title("Bar Plot", fontsize="medium")

bar_plt_data_normalized = df["col_2"].value_counts(normalize=True)
axs[1].pie(bar_plt_data_normalized.values, labels=bar_plt_data_normalized.index, autopct='%1.1f%%')
axs[1].set_title("Pie Plot", fontsize="medium")

plt.show()
````

### Bivariate Plotting For Num vs Num - Scatter Plot and Hexbin Plot

````python
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout="constrained")
fig.suptitle("Bivariate Plotting - Categorical Features")

axs[0].scatter(df["col_3"], df["col_4"])
axs[0].set_title("Scatter Plot", fontsize="medium")

hb1 = axs[1].hexbin(df["col_3"], df["col_4"])
axs[1].set_title("Hexbin Plot", fontsize="medium")
fig.colorbar(hb1, ax=axs[1])

hb2 = axs[2].hexbin(df["col_3"], df["col_4"], bins="log")
axs[2].set_title("Hexbin Plot with log Color Scale", fontsize="medium")
fig.colorbar(hb2, ax=axs[2])

plt.show()
````
